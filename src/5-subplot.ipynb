### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x ** 2
y4 = np.sqrt(x)

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Sine Wave')

# Second subplot
axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('Cosine Wave')

# Third subplot
axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Quadratic Function')

# Fourth subplot
axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('Square Root Function')

# Add overall labels
fig.suptitle('Subplot Layout Example', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
```
